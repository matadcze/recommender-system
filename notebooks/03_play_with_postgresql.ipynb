## Connect and query with pandas:

```python
import pandas as pd
from sqlalchemy import create_engine

# Connection string
engine = create_engine("postgresql://recommender:recommender@localhost:5432/recommender",
    connect_args={"options": "-csearch_path=movielens"}) # using schema

# Query directly into DataFrame
df = pd.read_sql("SELECT * FROM movies LIMIT 10", engine)
df.head()
```

## For larger datasets, use chunking

```python
chunks = pd.read_sql("SELECT * FROM movielens.ratings", engine, chunksize=10000000)

for chunk in chunks:
    # process each chunk
    print(chunk.shape)
```
