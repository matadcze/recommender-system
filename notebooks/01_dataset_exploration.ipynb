```python
import pandas as pd
import zipfile
from pathlib import Path
from kaggle.api.kaggle_api_extended import KaggleApi

api = KaggleApi()
api.authenticate()
```

```python
dataset = 'grouplens/movielens-20m-dataset'
DATA_PATH = Path("../data")
DATA_PATH.mkdir(parents=True, exist_ok=True)
```

```python
api.dataset_download_files(dataset, path=DATA_PATH, unzip=False )
```

```python
zip_path = DATA_PATH/"movielens-20m-dataset.zip"

def get_preview(file_name, n=100):
    with zipfile.ZipFile(zip_path, 'r') as z:
        with z.open(file_name) as f:
            return pd.read_csv(f, nrows=n)

df_movie = get_preview('movie.csv')
df_rating = get_preview('rating.csv')
df_link = get_preview('link.csv')
df_tag = get_preview('tag.csv')
df_genome_scores = get_preview('genome_scores.csv')
df_genome_tags = get_preview('genome_tags.csv')
```
